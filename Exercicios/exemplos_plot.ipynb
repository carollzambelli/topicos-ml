```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stock_data.txt")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date']) 
```

Gráfico de linha preço ao longo do tempo

```python
new_df = (df['Date']>=  pd.to_datetime('2020-4-1')) & (df['Date']<= pd.to_datetime('2020-09-30'))
df1 = df.loc[new_df]
df2 = df1.set_index('Date')
plt.figure(figsize=(10,5))
plt.suptitle('Stock prices', fontsize=18, color='black')
plt.xlabel("Date",fontsize=16, color='black')
plt.ylabel("$ price", fontsize=16, color='black')
 
df2['Close'].plot(color='green')
plt.show()
```

Gráfico de barras volume de negociação

```python
new_df = (df['Date']>= pd.to_datetime('2020-4-1')) & (df['Date']<= pd.to_datetime('2020-4-30')       )
df1 = df.loc[new_df]
df2 = df1.set_index('Date')
plt.figure(figsize=(10,6))
plt.suptitle('Trading Volume', fontsize=16, color='black')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("Trading Volume", fontsize=12, color='black') 
df2['Volume'].plot(kind='bar');
plt.show()
```

Stacked Bar

```python
new_df = (df['Date']>= pd.to_datetime('2020-4-1')) & (df['Date']<= pd.to_datetime('2020-4-30'))
df1 = df.loc[new_df]
df2 = df1.set_index('Date')
plt.figure(figsize=(15,10))
df2[['Open', 'Close']].plot.bar(stacked=True);
plt.suptitle('Opening/Closing stock prices', fontsize=12, color='black')
plt.show()
```

HIstograma

```python
new_df = (df['Date']>= pd.to_datetime('2020-4-1')) & (df['Date']<= pd.to_datetime('2020-9-30'))
df1 = df.loc[new_df]
df2 = df1['Open']
plt.figure(figsize=(10,6))
df2.plot.hist(alpha=0.5)
plt.suptitle('Opening stock Histogram', fontsize=12, color='blue')
plt.show()
```

Subplots

```python
new_df = (df['Date']>= pd.to_datetime('2020-4-1')) & (df['Date']<= pd.to_datetime('2020-9-30'))
df1 = df.loc[new_df]
stock_data = df1.set_index('Date')
stock_data.plot(subplots = True, figsize = (12, 8));
plt.legend(loc = 'best')
plt.suptitle('Open,High,Low,Close,Adj Close prices & Volume', fontsize=12, color='black')
plt.show()
```
